```python
def replace_data_spikes(row, max_counter): 
    counter = row["ENTRIES"] - row["PREV_ENTRIES"]
    if counter > max_counter:
        row["ENTRIES"] = row["PREV_ENTRIES"]
    return row["ENTRIES"]

mta_df["NEW_ENTRIES"] = mta_df.apply(replace_data_spikes, axis=1, max_counter=7500)
```

```python
mta_df[["PREV_DATE", "PREV_ENTRIES"]] = (mta_df.groupby(["CA", "UNIT", "SCP", "STATION"])["DATE", "NEW_ENTRIES"]
                                        .apply(lambda grp: grp.shift(1)))
```

```python
def get_daily_counts(row, max_counter):
    counter = row["NEW_ENTRIES"] - row["PREV_ENTRIES"]
    if counter > max_counter:
        return 0
    return counter

mta_df["HOURLY_ENTRIES"] = mta_df.apply(get_daily_counts, axis=1, max_counter=7500)
```

```python
def time_buckets(num):
    if num >= 2 and num < 6:
        return 'midnight to 4am'
    elif num >= 6 and num < 10:
        return '4am to 8am'
    elif num >= 10 and num < 14:
         return '8am to noon'
    elif num >= 14 and num < 18:
         return 'noon to 4pm'
    elif num >= 8 and num < 22:
        return '8pm to midnight'
    elif num >= 22 or num < 2:
        return 'midnight to 4am'
    else:
        return "xyz"
```

```python
mta_df["TIME_BUCKET"] = mta_df["HOUR"].apply(time_buckets)
```
